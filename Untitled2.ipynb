#4.Describe how dictionaries store data?
 -Dictionaries in Python are highly optimized hash table implementations. Here's how they work:
Core Structure
Dictionaries store data as key-value pairs, where:

Each key must be unique and immutable (strings, numbers, tuples of immutable elements)
Values can be any Python object (mutable or immutable)

Implementation Details

1.Hash Tables: Python dictionaries use a hash table data structure under the hood.
2.Hashing Process:

When you create a key-value pair, Python computes a hash value for the key using the hash() function
This hash value is used to determine where in memory the value should be stored
The hash value is mapped to a specific index in an array (called the hash table)


3.Collision Handling: Python uses open addressing with probing to resolve hash collisions (when two different keys produce the same hash value).
4.Dynamic Resizing: When a dictionary becomes too full, Python automatically resizes it to maintain efficiency:

Creates a larger hash table
Rehashes all existing key-value pairs into the new table



Performance Characteristics

Lookups: O(1) average case - extremely fast regardless of size
Insertions: O(1) average case
Deletions: O(1) average case

Memory Structure
A simplified view of the dictionary's internal structure:
{
  indices:   [0,     1,     2,      3,     4,    ...]
  entries:   [None, (k1,v1), None, (k2,v2), None, ...]
}
Example-
student = {
    "name": "Alice",
    "age": 21,
    "courses": ["Math", "Computer Science"]
}
In this example, "name", "age", and "courses" are hashed to determine their storage locations in the dictionary's underlying array.
Since Python 3.7+, dictionaries preserve insertion order, meaning when you iterate through a dictionary, you'll see keys in the same order they were added (this was previously an implementation detail, now it's guaranteed).

#8. What is a hash table, and how does it relate to dictionaries in Python?
 -Hash Tables and Python Dictionaries
Hash Tables: The Fundamental Concept
A hash table is a data structure that maps keys to values using a special function called a hash function. It works by:

Converting each key to a numeric hash value using a hash function
Using that hash value to determine where to store the key-value pair in memory
Providing nearly constant-time O(1) lookups, insertions, and deletions regardless of size

The main components of a hash table are:

An array (the "table") for storing data
A hash function that converts keys to array indices
A collision resolution strategy for when different keys hash to the same index

Python Dictionaries as Hash Tables
Python dictionaries are implemented as hash tables with some optimizations:

-1.Direct Implementation: Dictionaries in Python are hash tables - not just based on them, but actual optimized hash table implementations
-2.Hash Function: Python uses built-in hash() function to generate numerical values from keys
-3.Collision Resolution: Python uses open addressing with probing to handle collisions
-4.Dynamic Resizing: Dictionaries automatically resize when they get too full to maintain efficiency

How Python Dictionaries Work
When you create a dictionary:
student = {"name": "Alice", "age": 21}
Python:

Computes hash("name") and hash("age")
Uses these hash values to determine where in memory to store the values "Alice" and 21
When you access student["name"], it computes hash("name") again and goes directly to that memory location

This is why dictionary lookups are so fast - Python doesn't need to search through all items sequentially.
Key Requirements
For the hash table to work properly, dictionary keys must be:

Hashable: They must have a __hash__() method that returns a consistent value
Comparable: They must have an __eq__() method to determine equality
Immutable: Their hash value must not change over time

This is why lists and dictionaries can't be used as dictionary keys, but strings, numbers, and tuples (of hashable objects) can.
The hash table implementation is what gives Python dictionaries their remarkable performance characteristics and makes them such a powerful and widely used data structure.

#19.Why are dictionaries considered unordered, and how does that affect data retrieval?
-Why Dictionaries Were Historically Considered Unordered
In Python versions before 3.7, dictionaries were unordered collections. This meant:

The order in which you inserted key-value pairs was not guaranteed to be preserved.

When you iterated over the dictionary (e.g., with a for loop), the items could come out in any order.

This design was based on how dictionaries were implemented internally using hash tables, which prioritized fast lookup, insertion, and deletion rather than maintaining order.

What's Changed Since Python 3.7?
Starting with Python 3.7, the language guarantees that insertion order is preserved in dictionaries. So, if you do:

python
Copy
Edit
d = {'a': 1, 'b': 2, 'c': 3}
for key in d:
    print(key)
You’ll always get:

css
Copy
Edit
a
b
c
This order is preserved because Python now uses an order-preserving variant of a hash table internally.

Why Are They Still Conceptually Called Unordered?
Even though the implementation maintains order now, dictionaries are still conceptually unordered because:

Their primary purpose is fast key-based access, not sequence-like behavior.

You shouldn't rely on ordering for correctness unless explicitly required.

The term “unordered” is a reminder that you should use collections.OrderedDict (or lists) if your code semantically depends on order (especially for older Python versions).

How Does This Affect Data Retrieval?
Regardless of insertion order:

Retrieving values by key (my_dict[key]) is constant time on average—fast and unaffected by order.

If you loop over items, the order is now reliable (in 3.7+), so you can depend on it for things like consistent output, serialization, or testing.
